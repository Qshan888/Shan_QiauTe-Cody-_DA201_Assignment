### Objective 1
Create three visualisations indicating the number of appointments per month for service settings, context types, and national categories.

```python
# Change the data type of the appointment month to string to allow for easier plotting.
ar['appointment_month'] = ar['appointment_month'].astype('string')
nc['appointment_month'] = nc['appointment_month'].astype('string')

# Check for if appointment month is a string
print(ar['appointment_month'])
print(nc['appointment_month'])
```

```python
# Aggregate on monthly level and determine the sum of records per month.
nc_ss = nc.groupby([nc['appointment_month'],nc['service_setting']])['count_of_appointments'].agg(['count', 'mean', 'min', 'max','sum'])\
    .sort_values([('sum')], ascending=False)


# Reset index for dataframe
nc_ss = nc_ss.reset_index()

# View dataframe
nc_ss.head()
```

**Service settings:**

```python
# Change sum to float 64
nc_ss['sum'] = nc_ss['sum'].astype('float64')
nc_ss.dtypes
```

```python
# Plot the appointments over the available date range, and review the service settings for months.

# Create a lineplot.
ax1 = sns.lineplot(x='appointment_month',y='sum_apppointments',hue='service_setting',data=nc_ss, ci=None)

# Adjust y axis to Thousands K
ylabels = ['{:,.0f}'.format(y) + 'K' for y in ax1.get_yticks()/1000]

#Apply to y label
ax1.set_yticklabels(ylabels)

# Plot axis and define formatting
plt.legend(loc="upper right", frameon=True, fontsize=14)
ax1.set_title('National Service Setting by Appointments', fontsize = 20)
ax1.set_xlabel("Year-Months", fontsize = 20)
ax1.set_ylabel("Appointments", fontsize = 20)
ax1

# Assign subplot to a figure and save figure
fig = ax1.get_figure()
fig.savefig('Visual Graphs/National Service Setting by Appointments.png')
```

```python
# Load the data set.
tweets = pd.read_csv("Data/tweets.csv")

# set column width
pd.options.display.max_colwidth=200

# View the DataFrame.
tweets.head()
```

```python
# Change to a new dataframe with only text values
tweets_text = tweets.astype('string')

# Check if all columns are text values
tweets_text.info()
```

```python
# Create a list and append hastags into the list from the series

tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
# Select the first 30 tags
tags[0:29]
```

```python
tweets_text
```

```python
tweets.groupby(tweets['tweet_entities_hashtags'])['tweet_retweet_count'].agg(['count', 'mean', 'min', 'max','sum'])\
    .sort_values([('count')], ascending=False)
```

```python
# converst series into a dataframe
data = pd.DataFrame(tags, columns = ['word'])


# describe dataframe
data.describe()
```

```python
# add a column with the count of the word or tag
data['count'] = data.groupby('word')['word'].transform('count')
```

```python
# sort for highest counts
data = data.sort_values('count', ascending=False)


# view dataframe
data
```

```python
# remove any duplicates
data = data.drop_duplicates()

# create dataframe where count > 20
data_filtered = data[data['count'] > 25]

# view filtered dataframe
data_filtered
```

```python
# Plot and add data labels
ax9 = sns.barplot(x='word',y='count',data=data_filtered)

for i in ax9.containers:
    ax9.bar_label(i,)

# Format the axis and titles    
ax9.set_title('Top Trending Hashtags# on Twitter related to healthcare in UK', fontsize = 20)
ax9.set_xlabel("Hashtags", fontsize = 20)
ax9.set_ylabel("Count of Hashtag#", fontsize = 20)

# View the subplot
ax9

# Assign subplot to a figure and save figure
fig = ax9.get_figure()
fig.savefig('Visual Graphs/Top Trending Hashtags# on Twitter related to healthcare in UK.png')
```
